# SOTTO-FUNZIONI CALLABLE PER IL FRONT-END

tutte le funzioni qui sotto vanno precedute dalla decoration "@anvil.server.callable"

```python
@anvil.server.callable
def getGraphic( a , b, choice, cond1, cond2, cond3, nome ) :

  if choice == 0 :
    if a == "VENDITE":
      if b == "GLOBALE":
        table = pd.read_excel("result/VScostamenti_Vendite_Globali.xlsx")
      elif b == "CLIENTE":
        table = pd.read_excel("result/Scostamenti_VenditeC.xlsx") #Scostamenti_VenditeC
      elif b == "ARTICOLO":
        table = pd.read_excel("result/Scostamenti_VenditeA.xlsx") #Scostamenti_VenditeA

    elif a == "MP":
      table = pd.read_excel("result/Scostamenti_MateriePrime_Globali.xlsx")
    elif a == "MOD":
      table = pd.read_excel("result/Scostamenti_MOD_Globali.xlsx")
    
  elif choice == 1 :
       if b == "CLIENTE":
         table = pd.read_excel("result/Scostamenti_VenditeC.xlsx")
         table = table[ ( abs(table['ScostamentoVolumi']/table['BudgetStd']) > cond1) & ( abs(table['ScostamentoMix']/table['Mxstd'])  > cond2)  & ( abs(table['ScostamentoPrezzo']/table['Consuntivo']) > cond3 ) ]
       elif b == "ARTICOLO":
         table = pd.read_excel("result/Scostamenti_VenditeA.xlsx")
         table = table[ ( abs(table['ScostamentoVolumi']/table['BudgetStd']) > cond1) & ( abs(table['ScostamentoMix']/table['Mxstd'])  > cond2)  & ( abs(table['ScostamentoPrezzo']/table['Consuntivo']) > cond3 ) ]
    
  elif choice == 2 :
      query =  []
      query.append(nome)

      if b == "CLIENTE":
         tab = pd.read_excel("result/Scostamenti_VenditeC.xlsx")
         table = tab[tab['CodCliente'].isin(query) ]
      elif b == "ARTICOLO":
         tab = pd.read_excel("result/Scostamenti_VenditeA.xlsx")
         table = tab[tab['CodArticolo'].isin(query) ]
  
  table = table.drop('Unnamed: 0',1)

  return table.to_dict('list')

```
